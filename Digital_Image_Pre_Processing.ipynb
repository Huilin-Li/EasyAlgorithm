**A common digital image is an arrangment of pixels in $n$ dimensions**

**Each pixel has a specific value based on different bit depth**

<img src="pictures/bit_gray.png">

# 1. Libraries

```python
import matplotlib.pyplot as plt 
from skimage import io
import numpy as np
import seaborn as sns 
```

# 2. Load image.jpg

```python
img = io.imread("pictures/primary_color.jpg")
```

```python
#plt.axis('off')
plt.imshow(img)
```

```python
img.shape
```

# 3. Select a small square 

```python
square = img[330:360, 390:420, :]
plt.imshow(square)
```

```python
square.shape
```

# 4. Normalization of pixel values
#### Normalization can improve the numerical efficiency of image processing

```python
# the bit depth of the image
square.dtype
```

#### Therefore, this is 8-bit image.
#### The normalization method is dividing each pixel value by 255 .

```python
normalized_square = square/255.0
normalized_square.shape
```

# 5. Read the Red Channel

```python
normalized_square_R = normalized_square[:,:,0]
```

```python
plt.figure(figsize=(50,50)) 
sns.heatmap(normalized_square_R, annot= True, cmap='gray_r', cbar=False)
```

# 6. Read the Green Channel

```python
normalized_square_G = normalized_square[:,:,1]
```

```python
plt.figure(figsize=(50,50)) 
sns.heatmap(normalized_square_G, annot= True, cmap='gray_r', cbar=False)
```

# 7. Read the Blue channel

```python
normalized_square_B = normalized_square[:,:,2]
```

```python
plt.figure(figsize=(50,50)) 
sns.heatmap(normalized_square_B, annot= True, cmap='gray_r', cbar=False)
```

# 8. Pick one pixel, and have a look of its value in three channels

```python
pixel_x = 21
pixel_y = 21
```

```python
normalized_square_xy = normalized_square[pixel_y, pixel_x, :]
red_xy = normalized_square_R[pixel_y, pixel_x]
green_xy = normalized_square_G[pixel_y, pixel_x]
blue_xy = normalized_square_B[pixel_y, pixel_x]
```

```python
normalized_square_xy, red_xy, green_xy, blue_xy
```
